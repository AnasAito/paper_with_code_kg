```python
import pandas as pd
import ast
import hashlib 
base_path = '../data/bronze/'
```

```python
def generate_id(string):
    return hashlib.sha256(str(string).encode('utf-8')).hexdigest()
```

```python
# load data from bronze 
papers = pd.read_csv(base_path + 'papers.csv')
methods = pd.read_csv(base_path + 'methods.csv')
code= pd.read_csv(base_path + 'codes.csv')
evaluations = pd.read_csv(base_path + 'evaluation.csv')

```

# Nodes
## Papers Nodes 

```python
schema = {
    'id':'url_abs',
    'title':'title',
    'abstract':'abstract',
    'year':'date_year',
    'url' : 'url_abs',
    'pwc_id':'paper_url',
}
papers_nodes = papers[schema.values()]
papers_nodes.columns = list(schema.keys())
papers_nodes['label'] = 'paper'
papers_nodes
```

## Authors Nodes

```python
authors_2d = papers['authors'].values
all_authors = set()
for author_list  in authors_2d:

    try : 
            # string to list
        author_list = ast.literal_eval(author_list)
        for author in author_list:
            if author !='':
                all_authors.add(author)
    except:
        print(author_list)

all_authors = list(all_authors)

# create pandas dataframe with all authors
authors_nodes = pd.DataFrame(all_authors, columns=['name'])
```

```python
# create id from name hash 
authors_nodes['id'] = authors_nodes['name'].apply(generate_id)
authors_nodes['id'] = authors_nodes['id'].apply(lambda x: 'author_' + x)
authors_nodes['label'] = 'author'
```

## Tasks Nodes


```python
tasks_2d = papers['tasks'].values
all_tasks = set()
for tasks_list  in tasks_2d:

    try : 
            # string to list
        tasks_list = ast.literal_eval(tasks_list)
        for task in tasks_list:
            if task !='':
                all_tasks.add(task)
    except:
        print(tasks_list)

all_tasks = list(all_tasks)

# create pandas dataframe with all authors
tasks_nodes = pd.DataFrame(all_tasks, columns=['name'])
```

```python
# create id from name hash
tasks_nodes['id'] = tasks_nodes['name'].apply(generate_id)
tasks_nodes['id'] = tasks_nodes['id'].apply(lambda x: 'task_' + x)
tasks_nodes['label'] = 'task'
```

```python
tasks_nodes
```

## Evaluations Nodes 

```python
evaluations_nodes = evaluations
evaluations_nodes['string_id'] = evaluations_nodes['task']+'_'+evaluations_nodes['dataset']+'_'+evaluations_nodes['paper_url']
evaluations_nodes['id'] = evaluations_nodes['string_id'].apply(generate_id)
evaluations_nodes['id'] = evaluations_nodes['id'].apply(lambda x: 'evaluation_' + x)
evaluations_nodes['label'] = 'evaluation'

evaluations_nodes = evaluations_nodes[['id','model_name','metrics','label','paper_url','task','dataset']]
evaluations_nodes

```

## Datasets Nodes


```python
datasets_nodes = evaluations_nodes[['dataset']].drop_duplicates()

datasets_nodes['id'] = datasets_nodes['dataset'].apply(generate_id)
datasets_nodes['id'] = datasets_nodes['id'].apply(lambda x: 'dataset_' + x)
# rename column dataset to label

datasets_nodes['label'] = 'dataset'
datasets_nodes.columns = ['name','id','label']
datasets_nodes
```

## Methods Nodes


```python
methods_nodes = methods[['name','full_name','paper','introduced_year','url']]
methods_nodes['id'] = methods_nodes['url']
methods_nodes['id'] = methods_nodes['id'].apply(lambda x: 'method_' + x)
methods_nodes['label'] = 'method'
methods_nodes
```

## Codes Nodes

```python
codes_nodes = code[['repo_url','paper_url_abs','mentioned_in_paper','is_official','framework']]
codes_nodes['id'] = codes_nodes['repo_url']
codes_nodes['label'] = 'code'
codes_nodes
```

# Edges
#  author-> paper

```python
# create edges between authors and papers
authors_papers =[]
for paper in papers[['url_abs','authors']].values: 
    arxiv_id,authors = paper
    try : 
        authors = ast.literal_eval(authors)
        for author in authors:
            if author !='':
                authors_papers.append({'source':author, 'target':arxiv_id})
    except:
        print(authors)


# to pandas 
authors_papers = pd.DataFrame(authors_papers)
authors_papers.columns = ['source','target']

# join with authors_nodes
authors_nodes_ = authors_nodes
authors_nodes_['source'] = authors_nodes_['name']
authors_papers = authors_papers.merge(authors_nodes, on='source')
# rename 
authors_papers = authors_papers[['target','id']]
authors_papers.columns = ['target','source']
# edge label 
authors_papers['label'] = 'wrote'
```

## mehod -> paper

```python
method_paper =[]
for method in methods_nodes[['id','paper']].values : 
    try: 
        method_id,paper = method
        # string to dict 
        paper = ast.literal_eval(paper)
        paper_url = paper['url']
        method_paper.append({'source':method_id, 'target': paper_url})
    except:
       # print(method)
        pass
# to pandas 
method_paper = pd.DataFrame(method_paper)
method_paper.columns = ['source','target']
# join with papers_nodes
papers_nodes_ = papers_nodes
papers_nodes_['target'] = papers_nodes_['pwc_id']
method_paper = method_paper.merge(papers_nodes_, on='target')
# rename
method_paper = method_paper[['source','id']]
method_paper.columns = ['source','target']
# edge label
method_paper['label'] = 'first_mentioned_in'
```

## task -> paper

```python
# create edges between tasks and papers
tasks_papers =[]
for paper in papers[['url_abs','tasks']].values: 
    arxiv_id,tasks = paper
    try : 
        tasks = ast.literal_eval(tasks)
        for task in tasks:
            if task !='':
                tasks_papers.append({'source':task, 'target':arxiv_id})
    except:
        print(tasks)


# to pandas 
tasks_papers = pd.DataFrame(tasks_papers)
tasks_papers.columns = ['source','target']

# join with authors_nodes
tasks_nodes_ = tasks_nodes
tasks_nodes_['source'] = tasks_nodes_['name']
tasks_papers = tasks_papers.merge(tasks_nodes_, on='source')
# rename 
tasks_papers = tasks_papers[['target','id']]
tasks_papers.columns = ['target','source']
# edge label 
tasks_papers['label'] = 'benchmarked_in'
```

## paper -> evaluation

```python
evals_papers = evaluations[['id','paper_url']]
evals_papers.columns = ['source','target']
evals_papers['label'] = 'submitted'
```

## dataset -> evaluation


```python
evals_datasets = evaluations[['id','dataset']]
evals_datasets.columns = ['target','source']

# join with datasets_nodes
datasets_nodes_ = datasets_nodes[['id','name']]
datasets_nodes_['source'] = datasets_nodes_['name']
evals_datasets = evals_datasets.merge(datasets_nodes_, on='source')
# rename
evals_datasets = evals_datasets[['target','id']]
evals_datasets.columns = ['target','source']
# edge label
evals_datasets['label'] = 'used_in'
```

## task -> evaluation


```python
evals_tasks = evaluations[['id','task']]
evals_tasks.columns = ['target','source']

# join with tasks_nodes
tasks_nodes_ = tasks_nodes[['id','name']]
tasks_nodes_['source'] = tasks_nodes_['name']
evals_tasks = evals_tasks.merge(tasks_nodes_, on='source')
# rename
evals_tasks = evals_tasks[['target','id']]
evals_tasks.columns = ['target','source']
# edge label
evals_tasks['label'] = 'scored_'
```

## code -> paper 

```python
code_papers = code[['repo_url','paper_url_abs']]
code_papers.columns = ['source','target']
code_papers['label'] = 'mentioned_in'
```

# Aggregations

```python
graph = {
    'nodes': [authors_nodes, methods_nodes, papers_nodes, datasets_nodes, tasks_nodes,evaluations_nodes, codes_nodes],
    'edges': [authors_papers, method_paper, tasks_papers, evals_papers, evals_datasets, evals_tasks, code_papers]
}
# stats 
print('authors_nodes: ', f'{len(authors_nodes):,}')    
print('methods_nodes: ', f'{len(methods_nodes):,}')
print('papers_nodes: ', f'{len(papers_nodes):,}')
print('datasets_nodes: ', f'{len(datasets_nodes):,}')
print('tasks_nodes: ', f'{len(tasks_nodes):,}')
print('evaluations_nodes: ', f'{len(evaluations_nodes):,}')
print('codes_nodes: ', f'{len(codes_nodes):,}')
all_nodes_count = len(authors_nodes) + len(methods_nodes) + len(papers_nodes) + len(datasets_nodes) + len(tasks_nodes) + len(codes_nodes)
print('all_nodes_count: ', f'{all_nodes_count:,}')

print('-----')
print('authors_papers: ', f'{len(authors_papers):,}')
print('method_paper: ', f'{len(method_paper):,}')
print('tasks_papers: ', f'{len(tasks_papers):,}')
print('evals_papers: ', f'{len(evals_papers):,}')
print('evals_datasets: ', f'{len(evals_datasets):,}')
print('evals_tasks: ', f'{len(evals_tasks):,}')
print('code_papers: ', f'{len(code_papers):,}')
all_edges_count = len(authors_papers) + len(method_paper) + len(tasks_papers) + len(evals_papers) + len(evals_datasets) + len(evals_tasks) + len(code_papers)
print('all_edges_count: ', f'{all_edges_count:,}')

```
